```python
import pickle
```

```python
f0 = open('data0.pkl', 'rb')
f05 = open('data05.pkl', 'rb')
f1 = open('data1.pkl', 'rb')
dic0 = pickle.load(f0)
dic05 = pickle.load(f05)
dic1 = pickle.load(f1)
f0.close()
f05.close()
f1.close()
```

```python
import pandas as pd
import numpy as np
    
df0 = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in dic0.items() ]))
df05 = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in dic05.items() ]))
df1 = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in dic1.items() ]))

```

```python
df0_stats = pd.concat([df0.mean(), df0.std()], axis=1)
df05_stats = pd.concat([df05.mean(), df05.std()], axis=1)
df1_stats = pd.concat([df1.mean(), df1.std()], axis=1)
```

```python
df0_stats=df0_stats.rename(columns={0:'mean', 1:'std'})
df05_stats=df05_stats.rename(columns={0:'mean', 1:'std'})
df1_stats=df1_stats.rename(columns={0:'mean', 1:'std'})
```

```python
df0_stats.plot.bar(rot=0, yerr='std', legend=False, xlabel='season', ylabel='Average number of positive pixel', title='Average number of positive pixel (i.e. t=0)')
```

```python
df05_stats.plot.bar(rot=0, yerr='std', legend=False, xlabel='season', ylabel='Average number of positive pixel', title='t=0.5')
```

```python
df1_stats.plot.bar(rot=0, yerr='std', legend=False, xlabel='season', ylabel='Average number of positive pixel', title='t=1')
```
